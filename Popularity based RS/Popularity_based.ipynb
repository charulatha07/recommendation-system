```python
#read the ratings dataset
r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=r_cols,encoding='latin-1')
```

```python
ratings
```

```python
movie_ratings = pd.merge(items, ratings, left_on='movie id', right_on='movie_id')
```

```python
movie_ratings
```

```python
# Calculate popularity metrics
popularity = movie_ratings.groupby('movie id').agg(
    rating_count=('rating', 'count'),
    avg_rating=('rating', 'mean')
).reset_index()
```

```python
# Merge popularity data back to movies
movies_with_popularity = pd.merge(items, popularity, on='movie id')
```

```python
# List of genres from your dataset
genres = ['Action', 'Adventure', 'Animation', "Children's", 'Comedy', 'Crime',
          'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical',
          'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
```

```python
# Get top 5 per genre
top_movies_per_genre = {}

for genre in genres:
    top_movies = (
        movies_with_popularity[movies_with_popularity[genre] == 1]
        .sort_values(by='rating_count', ascending=False)
        .head(5)
        [['movie id', 'movie title', 'rating_count', 'avg_rating']]
    )
    top_movies_per_genre[genre] = top_movies
```

```python
# Display results
for genre, df in top_movies_per_genre.items():
    print(f"\nTop 5 {genre} movies:")
    print(df)
```

# combined table format 

```python
# Build the combined DataFrame
all_top_movies = []

for genre in genres:
    top_movies = (
        movies_with_popularity[movies_with_popularity[genre] == 1]
        .sort_values(by='rating_count', ascending=False)
        .head(5)
        [['movie id', 'movie title', 'rating_count', 'avg_rating']]
        .copy()
    )
    top_movies['genre'] = genre  # Add genre column
    all_top_movies.append(top_movies)
```

```python
# Combine into one DataFrame
final_df = pd.concat(all_top_movies, ignore_index=True)
```

```python
# Sort within genres
final_df = final_df.sort_values(by=['genre', 'rating_count'], ascending=[True, False])
```
